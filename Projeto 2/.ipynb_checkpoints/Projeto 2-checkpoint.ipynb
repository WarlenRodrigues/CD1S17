- A começar pela análise das médias de idade: há uma pequena evolução na média das expectativas de vida do ano de 1850 ao ano de 1924. No ano 2000 épossível verificar um salto bem grande na longevidade média das pessoas ao redor do mundo (o número médio de anos vividos quae dobrou).

A grande densidade de países localizadodos no interquartil da amostra justifica bem essa afimação. O país com maior longevidade (49,5 anos) faz parte de um grupo minoritário.

Por mais que a distriubuição dos dados possua uma assimetria positiva, os países com as maiores expectativas de vida (por mais que sejam minoria) ajudam a puxar a longevidade média para cima.
- O ano 2000 nos mostra que, além do salto observado na longevidade média, o desvio padrão se manteve semelhante ao de 1924. Tendo em mente os valores de máximo e mínimo do ano 2000, é possível inferir que os países não só aumentaram sua longevitade, como se tornaram melhor distribuídos. Ou seja, a assimetria na distribuição não é mais tão forte.

### Aderência de um modelo probabilístico normal aos dados

```python
f, (a1, a2, a3) = plt.subplots(3, figsize=(8,12))

a1.hist(anos.A1850, bins = 35, histtype='stepfilled',alpha=0.85, color='orange', normed =1)
a1.set_ylim(0.0, 0.15)
a1.set_xlabel("Expectativa de Vida (Anos)")
a1.set_ylabel("Densidade")
a1.set_title("1850")
x = sorted(stats.norm.rvs(size = 10000, loc = mi1850, scale = sigma1850))
a1.plot(x, stats.norm.pdf(x, loc = mi1850, scale = sigma1850), lw = 3, alpha = 0.6, color = 'Black')

a2.hist(anos.A1924, bins = 35, histtype='stepfilled',alpha=0.85, color='green', normed =1)
a2.set_ylim(0.0, 0.15)
a2.set_xlabel("Expectativa de Vida (Anos)")
a2.set_ylabel("Densidade")
a2.set_title("1924")
x = sorted(stats.norm.rvs(size = 10000, loc = mi1924, scale = sigma1924))
a2.plot(x, stats.norm.pdf(x, loc = mi1924, scale = sigma1924), lw = 3, alpha = 0.6, color = 'Black')

a3.hist(anos.A2000, bins = 35, histtype='stepfilled',alpha=0.85, color='cyan', normed =1)
a3.set_ylim(0.0, 0.12)
a3.set_xlabel("Expectativa de Vida (Anos)")
a3.set_ylabel("Densidade")
a3.set_title("2000")
x = sorted(stats.norm.rvs(size = 10000, loc = mi2000, scale = sigma2000))
a3.plot(x, stats.norm.pdf(x, loc = mi2000, scale = sigma2000), lw = 3, alpha = 0.6, color = 'Black')

plt.tight_layout(1)
```

```python
fempirica1850 = []
for i in range(len(anos.A1850)):
    fempirica1850.append ((i+0.5)/(len(anos.A1850)))
    
fempirica1924 = []
for i in range(len(anos.A1924)):
    fempirica1924.append ((i+0.5)/(len(anos.A1924)))

fempirica2000 = []
for i in range(len(anos.A2000)):
    fempirica2000.append ((i+0.5)/(len(anos.A2000)))


fnormal1850 = []
for i in anos.A1850:
    fnormal1850.append(stats.norm.cdf (i, loc = mi1850, scale = sigma1850))

fnormal1924 = []
for i in anos.A1924:
    fnormal1924.append(stats.norm.cdf (i, loc = mi1924, scale = sigma1924))

fnormal2000 = []
for i in anos.A2000:
    fnormal2000.append(stats.norm.cdf (i, loc = mi2000, scale = sigma2000))
```

```python
f, (comp1, comp2, comp3) = plt.subplots(3, figsize=(8,12))

comp1.plot(anos.A1850, fnormal1850, color='Black', lw=1.3, label='Frequência através da Normal')
comp1.plot(anos.A1850, fempirica1850, 'o',alpha=0.85, color='red', label = 'Frequência através de dados Empíricos')
comp1.set_xlabel("Valores Amostrais da Longevidade em 1850")
comp1.set_ylabel("Frequência Relativa Acumulada")
comp1.grid(True)
comp1.legend (loc = 'best')


comp2.plot(anos.A1924, fnormal1924, color='Black', lw=1.3, label='Frequência através da Normal')
comp2.plot(anos.A1924, fempirica1924, 'o',alpha=0.85, color='yellow', label = 'Frequência através de dados Empíricos')
comp2.set_xlabel("Valores Amostrais da Longevidade em 2000")
comp2.set_ylabel("Frequência Relativa Acumulada")
comp2.grid(True)
comp2.legend (loc = 'best')


comp3.plot(anos.A2000, fnormal2000, color='Black', lw=1.3, label='Frequência através da Normal')
comp3.plot(anos.A2000, fempirica2000, 'o',alpha=0.85, color='blue',  label = 'Frequência através de dados Empíricos')
comp3.set_xlabel("Valores Amostrais da Longevidade em 2000")
comp3.set_ylabel("Frequência Relativa Acumulada")
comp3.grid(True)
comp3.legend (loc = 'best')

plt.tight_layout(1.5)
```

Através dos gráficos de aderência é possível notar que o ano de 1850, por mais que não caiba no modelo Normal no que diz respeito ao histograma, se encaixa bem na Função de Densidade Acumulada da Normal com a respectiva média e desvio padrão.

```python
quantil1850 = []
for i in fempirica1850:
    quantil1850.append(stats.norm.ppf (i, loc = mi1850, scale = sigma1850))

quantil1924 = []
for i in fempirica1924:
    quantil1924.append(stats.norm.ppf (i, loc = mi1924, scale = sigma1924))
    
quantil2000 = []
for i in fempirica2000:
    quantil2000.append(stats.norm.ppf (i, loc = mi1850, scale = sigma1850))
```

```python
f, (comp1, comp2, comp3) = plt.subplots(3, figsize=(8,12))

comp1.plot(anos.A1850, quantil1850, color='green', lw=2)
comp1.plot ([10, 60], [10, 60], label = 'Bissetriz', color = 'Black', alpha = 1, lw=1)
comp1.set_xlabel("Quantil Amostral")
comp1.set_ylabel("Quantil Teórico")
comp1.grid(True)
comp1.legend (loc = 'best')


comp2.plot(anos.A1924, quantil1924, color='red', lw=2)
comp2.plot ([15, 80], [15, 80], label = 'Bissetriz', color = 'Black', alpha = 1, lw=1)
comp2.set_xlabel("Quantil Amostral")
comp2.set_ylabel("Quantil Teórico")
comp2.grid(True)
comp2.legend (loc = 'best')


comp3.plot(anos.A2000, quantil2000, color='yellow', lw=2)
comp3.plot ([10, 90], [10, 90], label = 'Bissetriz', color = 'Black', alpha = 1, lw=1)
comp3.set_xlabel("Quantil Amostral")
comp3.set_ylabel("Quantil Teórico")
comp3.grid(True)
comp3.legend (loc = 'best')

plt.tight_layout(1.5)
```

### Cálculo de probabilidades a partir da distribuição normal

```python
#1850
prop_70anos_1850 = 1-stats.norm.cdf(70, loc=mi1850, scale=sigma1850)
#1924
prop_70anos_1924 = 1-stats.norm.cdf(70, loc=mi1924, scale=sigma1924)
#2000
prop_70anos_2000 = 1-stats.norm.cdf(70, loc=mi2000, scale=sigma2000)
```

Probabilidade de um país ter longevidade maior que 70 anos de idade:

```python
print ('Para 1850:', prop_70anos_1850*100, '%')
print ('Para 1924:', prop_70anos_1924*100, '%')
print ('Para 2000:', prop_70anos_2000*100, '%')
```

Sabendo que um país possui expectativa maior que 70 anos, qual é a probabilidade de possuir menos de 75 anos de longevidade?

```python
prob_1850_75de70 = (1 - stats.norm.cdf(75, loc= mi1850, scale= sigma1850))/(1 - stats.norm.cdf(70, loc= mi1850,
                                                                                               scale= sigma1850))
prob_1924_75de70 = (1 - stats.norm.cdf(75, loc= mi1924, scale= sigma1924))/(1 - stats.norm.cdf(70, loc= mi1924,
                                                                                               scale= sigma1924))
prob_2000_75de70 = (1 - stats.norm.cdf(75, loc= mi2000, scale= sigma2000))/(1 - stats.norm.cdf(70, loc= mi2000,
                                                                                               scale= sigma2000))
```

```python
print ('Para 1850:', prob_1850_75de70*100, '%')
print ('Para 1924:', prob_1924_75de70*100, '%')
print ('Para 2000:', prob_2000_75de70*100, '%')
```

Qual a maior expectativa de vida dos 10% de países com menores expectativas de vida?

Para encontrarmos a maior longevidade dos 10% de países com menor expectativa de vida de cada ano temos que encontrar $z$ que retorne o valor mais próximo de 0.1. Ou seja: $z=\frac{x-\mu}{\sigma}$ = 0.1

$z$= 0.25

Para o ano de 1850:

$0.25=\frac{x-31.63}{4.59}$

$x$ = 32,7 anos


Para o ano de 1924:

$0.25=\frac{x-35.76}{9.31}$


$x$ = 38.0 anos

Para o ano de 2000:

$0.25=\frac{x-68.21}{9.01}$


$x$ = 70,4 anos

Para encontrarmos a menor longevidade dos 10% de países com maior expectativa de vida de cada ano temos que encontrar $z$ que retorne o valor mais próximo de 0.9. Ou seja: $z=\frac{x-\mu}{\sigma}$ = 0.9
