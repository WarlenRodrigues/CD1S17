```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
from scipy import stats

print('Esperamos trabalhar no diretório')
print(os.getcwd())
```

```python
tab = pd.read_excel('Life.xlsx', sep= ',')
anos = pd.DataFrame()
anos["A1850"] = tab.A1850
anos["A1924"] = tab.A1924
anos["A2000"] = tab.A2000

anos = anos.dropna(subset = ["A1850", "A1924", "A2000"])

anos.head()
```

___
## Seleção dos três anos com diferentes formatos de distribuição da expectativa de vida

```python
a11850 = sorted(anos.A1850)
mi1850 = anos.A1850.mean()
sigma1850 = (anos.A1850.var())**0.5

mi1924 = anos.A1924.mean()
sigma1924 = (anos.A1924.var())**0.5

mi2000 = anos.A2000.mean()
sigma2000 = (anos.A2000.var())**0.5


f, (a1, a2, a3) = plt.subplots(3, figsize=(8,12))

a1.hist(anos.A1850, bins = 35, histtype='stepfilled',alpha=0.85, color='orange', normed =1)
a1.set_ylim(0.0, 0.13)
a1.set_xlabel("Expectativa de Vida (Anos)")
a1.set_ylabel("Densidade")
a1.set_title("1850")
a1.plot(anos.A1850, stats.norm.pdf(anos.A1850, loc=mi1850, scale=sigma1850),alpha=1, lw=3, color='blue')

a2.hist(anos.A1924, bins = 35, histtype='stepfilled',alpha=0.85, color='green', normed =1)
a2.set_ylim(0.0, 0.15)
a2.set_xlabel("Expectativa de Vida (Anos)")
a2.set_ylabel("Densidade")
a2.set_title("1924")
a2.plot(anos.A1924, stats.norm.pdf(anos.A1924, loc=mi1924, scale=sigma1924),alpha=1, lw=3, color='blue')

a3.hist(anos.A2000, bins = 35, histtype='stepfilled',alpha=0.85, color='cyan', normed =1)
a3.set_ylim(0.0, 0.1)
a3.set_xlabel("Expectativa de Vida (Anos)")
a3.set_ylabel("Densidade")
a3.set_title("2000")
a3.plot(anos.A2000, stats.norm.pdf(anos.A2000, loc=mi2000, scale=sigma2000),alpha=1, lw=3, color='blue')

plt.tight_layout(1)

#anos["A1850"].hist(bins=35)
```
